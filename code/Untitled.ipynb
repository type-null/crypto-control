```python
def calculateReturn(df, coin: str):
    df.loc[:, "price"] = df.loc[:, coin]
    df.loc[:, "logPrice"] = np.log(df.loc[:, "price"])
    df.loc[:, "return"] = df.loc[:, "logPrice"] - df.loc[:, "logPrice"].shift(1)
    
    df.loc[df.index[0], "return"] = 0
    df.loc[df.index[0], 'signal'] = 0
    cols = ['price', 'logPrice', 'return', 'signal']
    return df.loc[:, cols]

#estimate mu0 and sigma for all of our coins
lambdaa=0.3 #by papers
L=3 # L=2 check later
mu0=[]
sigma=[]
for i in range(0,len(cryptos.columns)):
    mu0.append(mean(cryptos[cryptos.columns[i]]))
    sigma.append(np.std(cryptos[cryptos.columns[i]]))

def detectBuySignal(df, start, feature):
    
    #df.loc[df.index[start],'Z']=lambdaa*cryptos['BTC'].values[0]+(1-lambdaa)*mu0[0]
    #df.loc[df.index[start],'ucl']=mu0[0]+(L*sigma[0]*sqrt(lambdaa/((2-lambdaa)*1)*(1-(1-lambdaa)**(2*1))))
    #df.loc[df.index[start],'lcl']=mu0[0]-(L*sigma[0]*sqrt(lambdaa/((2-lambdaa)*1)*(1-(1-lambdaa)**(2*1))))
    Z = [lambdaa*cryptos['BTC'].values[0]+(1-lambdaa)*mu0[0]]
    ucl = [mu0[0]+(L*sigma[0]*sqrt(lambdaa/((2-lambdaa)*1)*(1-(1-lambdaa)**(2*1))))]
    lcl = [mu0[0]-(L*sigma[0]*sqrt(lambdaa/((2-lambdaa)*1)*(1-(1-lambdaa)**(2*1))))]
    for j,k in zip(range(start+1, len(df)), df.price[1:len(cryptos['BTC'])]):
        Z = lambdaa*k+(1-lambdaa)*Z[j-1]
        ucl=mu0[0]+(L*sigma[0]*sqrt(lambdaa/((2-lambdaa)*1)*(1-(1-lambdaa)**(2*(j+1)))))
        lcl=mu0[0]-(L*sigma[0]*sqrt(lambdaa/((2-lambdaa)*1)*(1-(1-lambdaa)**(2*(j+1)))))
        
        df.loc[df.index[j], 'Z']=Z
        df.loc[df.index[j], 'ucl']=ucl
        df.loc[df.index[j], 'lcl']=lcl
    
        if Z >= ucl:
            #buy signal
            df.loc[df.index[j], 'signal'] = 1
            if Z <= lcl:
                df.loc[df.index[j], 'signal'] = -1
            #retutn detectBuySignal(df, j)
        else:
            df.loc[df.index[j], 'signal'] = 0
    return df

#def detectBuySignal(df, start):
def createTradeSignal(df, feature='return'):
    return detectBuySignal(df, start=0, feature=feature)
```
